# 모든 것은 객체
- 파이썬 언어의 중요한 특징중 하나는 객체 모델의 일관성임

- 각 객체는 연관된 자료형과 내부 데이터를 가짐(심지어 함수마저도 하나의 객체로 간주함)

# 함수와 객체 매서드
- 함수는 괄호와 0개이상의 인수를 전달해서 호출할 수 있음. 반환되는 값은 선택적으로 변수에 대입할 수 있음

# 속성과 메서드

- as 예약어를 사용하면 모듈을 다른 이름으로 import할 수 있음


```python
%%writefile some_module.py
# 모듈 테스트용 파일
PI = 3.14159

def f(x):
    return x + 2

def g (a, b):
    return a + b
```

# 이항 연산자
1. a + b : a와 b를 더함
2. a - b : a에서 b를 뺀다
3. a * b : a와 b를 곱한다
4. a / b : a를 b로 나눈다
5. a // b : a를 b로 나눈 몫을 취한다(소수점 이하는 버림)
6. a ** b : a의 b 승을 구한다
7. a & b : a와 b 모두 True인 경우 True를 반환한다. 정수인 경우 비트 단위 AND를 구한다
8. a | b : a 혹은 b가 True인 경우 True를 반환한다. 정수인 경우 비트 단위 OR을 구한다
9. a ^ b : 불리언의 경우 a 혹은 b 중 하나만 True인 경우 True를 반환함. 정수인 경우 비트 단위 EXCLUSIVE-OR를 구한다
10. a == b : a와 b가 같은 경우 True
11. a != b : a와 b가 다른경우 True 
12. a < b, a <= b : a가 b보다 작은 경우, a가 b보다 작거나 같은 경우
13. a > b, a >= : a가 b보다 큰 경우, a가 b보다 작거나 큰 경우
14. a is b : a와 b가 같은 파이썬 객체를 참조할 경우 True
15. a is not b : a와 b가 다른 파이썬 객체를 참조할 경우 True

```python
# 두 참조 변수가 동일한 객체를 참조하는지 검사 하려면 is 예약어를 사용. is not을 사용하면 두 객체가 같지 않으닞 검사 할 수 있음.
a = [1, 2, 3]
b = a
c = list(a) # list는 항상 새로운 파이썬 리스트를 생성하므로 c는 a와 구별됨
a is b
```

```python
a is not c
```

```python
# is 와 is not은 변수가 None 인지 검사하기 워해 흔히 사용함. None 인스턴스는 하나만 존재하기 때문임
a = None
a is None
```

## 2.2.2 스칼라 자료형
- 파이썬은 숫자 데이터와 문자열, 불리언 값 그리고 날짜와 시간을 다룰 수 있는 몇몇 내장 자료형을 제공함
- 이런 단일 값을 담는 타입을 스칼라 자료형이라고 부름

# 표준 파이썬 스칼라 자료형
1. None : 파이썬의 null 값(하나의 유일한 None 인스턴스만 존재)
2. str : 문자열 자료형, 유니코드(UTF-8 인코딩) 문자열
3. bytes : 원시 이진 데이터
4. float : 배정밀도(64비트) 부동소수점수, double형이 따로 존재하지 않음
5. bool : 참(True) 또는 거짓(False)
6. int : 부호가 있는(음수 표현이 가능한) 정수, 값의 범위는 플랫폼에 의존적임

# 숫자 자료형
- 파이썬에서 숫자를 위한 주요 자료형은 int와 float
- int는 임의의 숫자를 자정할 수 있음
- 부동소수점 숫자는 float 자료형으로 나타냄, 내부적 배정밀도(64비트)를 가짐, 과학적 표기법으로 나타낼 수도 있음
- 정수 나눗셈은 int를 반환하지 않고 float를 반환함
- //는 소수부분을 삭제하고 몫만 돌려주는 연산자

```python
# 나눗셈 예시
3/2 
```

```python
3//2
```

# 문자열
