## 2.3 자기관찰
- 변수 이름 앞이나 뒤에 물음표(?)를 붙이면 그 객체에 대한 일반 정보를 출력함

```python
b = [1, 2, 3]
b? # 변수 b에 대한 정보 출력
```

```python
print? # 함수 print에 대한 정보 출력
```

- 자기관찰 시 객체가 함수이거나 인스턴스 매서드라면 정의 되어있는 문서(독스트링이)가 출력

```python
def add_numbers(a,b):
    """
    Add two numbers together
    
    Returns
    -------
    the_sum : type of arguments
    """
    return a + b


add_numbers? # 정의한 함수도 자기관찰 가능, 정의해준 문서도 출력 함 
```
